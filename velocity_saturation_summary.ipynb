
# Velocity Saturation in Semiconductors

## Average Thermal Energy at T = 300 K

The average thermal energy of a carrier at room temperature is given by:

$$
\frac{1}{2}mv_{th}^2 = \frac{3}{2}kT
$$

Where:
- $k = 8.617 \times 10^{-5}\ \text{eV/K}$ (Boltzmann constant)
- $T = 300\ \text{K}$

Thus:

$$
\frac{3}{2}kT = \frac{3}{2}(0.0259) = 0.03885\ \text{eV}
$$

---

## Mean Thermal Velocity

For an electron, this energy corresponds to a thermal velocity $v_{th} \approx 10^7\ \text{cm/s}$.

---

## Drift Velocity and Saturation Electric Field

Given:
- Electron mobility $\mu_n = 1350\ \text{cm}^2/\text{V·s}$
- Desired drift velocity $v_d = 10^5\ \text{cm/s}$

From the drift velocity equation:

$$
v_d = \mu_n E
$$

We compute the required electric field:

$$
E = \frac{v_d}{\mu_n} = \frac{10^5}{1350} \approx 74.07\ \text{V/cm}
$$

---

## Conclusion

An electric field of approximately **75 V/cm** results in a drift velocity of $10^5\ \text{cm/s}$, which is **1% of the thermal velocity** of an electron. At this field, the **velocity saturation effect** becomes relevant.

---

## Mobility Table at T = 300 K (Low Doping)

| Material         | $\mu_n$ (cm²/V·s) | $\mu_p$ (cm²/V·s) |
|------------------|------------------:|------------------:|
| Silicon          | 1350              | 480               |
| Gallium arsenide | 8500              | 400               |
| Germanium        | 3900              | 1900              |

---
## Nonlinear effect in Silicon are provided be following equations 

- **For electrons**:
  $$
  v_n = \frac{v_s}{\sqrt{1 + \left(\frac{E_{on}}{E}\right)^2}}
  $$

- **For holes**:
  $$
  v_p = \frac{v_s}{\sqrt{1 + \left(\frac{E_{op}}{E}\right)^2}}
  $$

## Constants:

- $v_s = 10^7\,\text{cm/s}$ (saturation velocity)
- $E_{on} = 7 \times 10^3\,\text{V/cm}$ (for electrons)
- $E_{op} = 2 \times 10^4\,\text{V/cm}$ (for holes)

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
vs = 1e7  # Saturation velocity in cm/s
Eon = 7e3  # Electric field constant for electrons (V/cm)
Eop = 2e4  # Electric field constant for holes (V/cm)

# Electric field range (log scale)
E = np.logspace(2, 6, 400)

# Drift velocity equations
vn = vs / np.sqrt(1 + (Eon / E) ** 2)
vp = vs / np.sqrt(1 + (Eop / E) ** 2)

# Plot
plt.figure(figsize=(8, 6))
plt.loglog(E, vn, label='Electrons', linewidth=2)
plt.loglog(E, vp, '--', label='Holes', linewidth=2)
plt.xlabel('Electric field (V/cm)', fontsize=12)
plt.ylabel('Drift Velocity (cm/s)', fontsize=12)
plt.title('Drift Velocity vs Electric Field in Silicon at T = 300 K', fontsize=14)
plt.grid(True, which="both", ls="--")
plt.legend()
plt.tight_layout()
plt.show()
```
