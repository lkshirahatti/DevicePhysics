# Semiconductor Electron Affinity (χ) + Bandgap (Eg) — vacuum level referenced

**Definition (vacuum level reference):**  
Electron affinity is defined as **χ = E_vac − E_CBM**, i.e., the energy difference between the vacuum level and the **conduction band minimum**.

**Important caveat:** χ values are **surface-dependent** (termination, orientation, adsorbates, oxides) and can shift by **~0.5–2 eV** in real devices.  
For alloys, χ and Eg vary with composition (the table lists common/representative values).
