# Doped Silicon Resistivity

This notebook computes and plots the **resistivity** of silicon at 300 K for both n-type and p-type doping using the formula: (Without non-linear mobility effects

$$
\rho = \frac{1}{q \mu N}
$$

Where:
- \( $\rho$ \) is resistivity (\($\Omega$\)-cm)
- \( q = 1.602 x $10^{-19}$ \) C
- \($\mu$ \) is the carrier mobility in \(${cm}^2$/Vs\)
- \( N \) is the impurity concentration in \(${cm}^{-3}$\)


```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
q = 1.602e-19  # Charge of electron in Coulombs

# Mobilities in cm^2/Vs
mu_n = 1350  # n-type mobility (electrons)
mu_p = 480   # p-type mobility (holes)

# Impurity concentration range (cm^-3)
N = np.logspace(12, 21, 1000)

# Resistivity calculations using ρ = 1 / (q * μ * N)
rho_n = 1 / (q * mu_n * N)
rho_p = 1 / (q * mu_p * N)

# Plot
plt.figure(figsize=(10, 7))
plt.loglog(N, rho_n, label='n type (phosphorus)', linewidth=2)
plt.loglog(N, rho_p, label='p type (boron)', linewidth=2)

plt.xlabel("Impurity concentration (cm$^{-3}$)", fontsize=12)
plt.ylabel("Resistivity ($\Omega$-cm)", fontsize=12)
plt.title("Resistivity vs Impurity Concentration for Silicon at 300 K", fontsize=14)
plt.grid(True, which='both', linestyle='-', linewidth=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
