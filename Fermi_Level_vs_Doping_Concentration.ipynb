# Fermi Level vs Doping Concentration at T = 300 K
This notebook calculates and plots the position of the Fermi level in silicon as a function of donor and acceptor doping concentrations.

```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
k = 8.617333262145e-5  # eV/K
T = 300  # K
kT = k * T

# Silicon band parameters
Eg = 1.12  # eV
Ec = Eg / 2  # conduction band edge (set as reference midpoint)
Ev = -Eg / 2  # valence band edge
Efi = 0.0  # intrinsic Fermi level at midgap (reference)
Nc = 2.8e19  # cm^-3
Nv = 1.04e19  # cm^-3

# Donor and acceptor concentrations (log scale range)
Nd = np.logspace(12, 18, 100)
Na = np.logspace(12, 18, 100)

# Fermi level expressions
Ef_n_type = Ec - kT * np.log(Nc / Nd)
Ef_p_type = Ev + kT * np.log(Nv / Na)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(Nd, Ef_n_type, label='n type', color='blue')
plt.plot(Na, Ef_p_type, label='p type', color='green')
plt.axhline(Efi, color='black', linestyle='--', label='Intrinsic $E_{Fi}$')
plt.axhline(Ec, color='blue', linestyle=':', linewidth=1, label='$E_c$')
plt.axhline(Ev, color='green', linestyle=':', linewidth=1, label='$E_v$')

plt.xscale('log')
plt.xlabel('Doping Concentration $N_d$ or $N_a$ (cm$^{-3}$)', fontsize=12)
plt.ylabel('Fermi Level Position (eV)', fontsize=12)
plt.title('Fermi Level vs Doping Concentration at T = 300 K', fontsize=14)
plt.grid(True, which='both', ls='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
