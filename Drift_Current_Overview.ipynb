
## Drift Current Equations

The **drift current density** is the current caused by charge carriers (electrons and holes) due to an electric field.

### For Holes:
$$
J_{p,\text{drf}} = (ep)v_{dp} = e \mu_p p E 
$$

### For Electrons:
The average drift velocity of an electron under an electric field:
$$
v_{dn} = -\mu_n E 
$$

So the electron drift current density becomes:
$$
J_{n,\text{drf}} = (-en)(-\mu_n E) = e \mu_n n E 
$$

**Note**: Although electrons move opposite to the electric field, the drift current direction is the same as the applied electric field.

### Total Drift Current Density:
Since both electrons and holes contribute to drift current:
$$
J_{\text{drf}} = e(\mu_n n + \mu_p p)E 
$$

## Mobility Effects

From Newton’s second law:

$$
F = m^* \frac{dv}{dt} = eE
$$

Assuming constant electric field:

$$
v = \frac{eEt}{m^*}
$$

Using average time between collisions \( $ \tau $ \):

$$
v_{\text{peak}} = \frac{e \tau}{m^*} E
$$

Average drift velocity:

$$
\langle v_d \rangle = \frac{1}{2} \left(\frac{e \tau}{m^*}\right) E
$$

Hole mobility:

$$
\mu_p = \frac{v_{dp}}{E} = \frac{e \tau_p}{m^*_p}
$$

Electron mobility:

$$
\mu_n = \frac{e \tau_n}{m^*_n}
$$


## Variable Definitions

- $J$: Drift current density (A/cm²)  
- $e$: Electron charge = $1.602 \times 10^{-19}$ C  
- $n$, $p$: Electron and hole concentration (cm⁻³)  
- $\mu_n$, $\mu_p$: Electron and hole mobility (cm²/V·s)  
- $E$: Electric field (V/cm)  
- $v_{dn}$, $v_{dp}$: Average drift velocities for electrons and holes (cm/s)
- $\tau_n$, $\tau_p$: Average time between collisions for electrons and holes

Thermal vibrations cause a disruption of the potential function, resulting in an interaction between the electrons or holes and the vibrating lattice atoms. This lattice scattering is also referred to as phonon scattering.
The second interaction mechanism affecting carrier mobility is called ionized impurity scattering. 

Mobility values are strong functions of the doping concentration and temperature. These variations must be taken into account in the design of semiconductor devices.
